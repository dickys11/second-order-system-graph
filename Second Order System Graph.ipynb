```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
# Values
Wn = 2*np.pi
zeta = [0, 1.0, 2.0]
```

```python
# Transfer Function
#  Wn**2 / (s**2 + 2 * zeta * Wn * s + Wn**2)
for z in zeta:
    num = [Wn*Wn]
    den = [1,2*z*Wn,Wn**2]
    sys1 = signal.TransferFunction(num,den)
    t1,y1 = signal.step(sys1)
    
    if z > 1.0:
        condition = 'Overdamped'
    elif 0 <= z < 1.0:
        condition = 'Underdamped'
    elif z == 1.0:
        condition = "Critically Damped"
    
    # Plot Image
    plt.plot(t1,y1,linewidth=3,label='zeta ='+ str(z))
    plt.title(condition + ', $\\omega_n = 2\\pi$, ')
    plt.xlabel('Time')
    plt.ylabel('Response (y)')
    plt.legend(loc='best')
    plt.show()
```
